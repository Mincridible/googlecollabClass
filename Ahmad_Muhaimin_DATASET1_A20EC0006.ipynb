## QUESTION 1 [10 MARKS]


Given numbers = [2,4,6,8,10,12]

(i) Find the average of the numbers, $\bar{x}$.

(ii) Then write a program that computes

$$s = \sqrt{\frac{1}{N-1} \sum_{i=1}^N (x_i - \overline{x})^2}$$

Use only Python basics (NumPy or pandas are not allowed).

```python
# WRITE YOUR CODE HERE
numbers = [2,4,6,8,10,12]
avg = sum(numbers)/len(numbers)
print(avg)

def maths(num):
    return ((1/len(numbers)-1)*((num-avg)**2))**(1/2)

nums = []

for i in numbers:
    nums.append(maths(i))
    
print(nums)
```

## QUESTION 2 [10 MARKS]

Write a function named `Weather` to performs these tasks

(i)&emsp;read temperature in centigrade, $C$ <br>
(ii)&emsp;display the temperature in fahrenheit, $F$ (formula to convert: $F = (C*1.8) + 32$) <br>
(iii)&emsp;create a nested conditional logic to display suitable message according to temperature state below:

    Temperature < 32 then Freezing
    Temperature 32-50 then Very Cold
    Temperature 51-68 then Cold
    Temperature 69-86 then Normal
    Temperature 87-104 then Hot
    Temperature >=105 then Very Hot


```python
# WRITE YOUR CODE HERE
def Weather(cels):
    msg=''
    fah = (cels*1.8)+32
    m1 = 'Temperature in Fahrenheit is: {:.2f}'.format(fah)
    if fah < 32:
        msg = 'The weather is Freezing'
    elif fah <= 50:
        msg = 'The weather is Very Cold'
    elif fah <= 68:
        msg = 'The weather is Cold'
    elif fah <= 86:
        msg = 'The weather is Normal'
    elif fah <= 104:
        msg = 'The weather is Hot'
    else:
        msg = 'The weather is Very Hot'
    
    return m1 +'\n'+ msg
    
```

```python
print(Weather(-10))
print(Weather(0))
print(Weather(15))
print(Weather(24))
print(Weather(37))
print(Weather(50))
```
